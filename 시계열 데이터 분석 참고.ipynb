```python
import pandas as pd
import pandas_datareader.data as web
import numpy as np
import matplotlib.pyplot as plt
import warnings
import pystan

from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
from fbprophet import Prophet
from datetime import datetime

warnings.simplefilter(action='ignore')
```

```python
url = 'https://finance.yahoo.com/quote/035420.KS/history?p=035420.KS&guccounter=1'

req = Request(url, headers={'User-Agent':'Chrome'})
page = urlopen(req).read()
soup = BeautifulSoup(page, 'html.parser')
table = soup.find('table')

df_raw = pd.read_html(str(table))[0]
df_raw # 맨 아래 값은 제외
```

```python
df_raw = df_raw.iloc[:-1, :]
df_raw.info()
```

```python
# date 형변환
df_raw['Date'] = pd.to_datetime(df_raw['Date'], format='%b %d, %Y')
# float형으로 형변환
df_raw['Open'] = pd.to_numeric(df_raw['Open'],errors = 'coerce')
df_raw['High'] = pd.to_numeric(df_raw['High'],errors = 'coerce')
df_raw['Low'] = pd.to_numeric(df_raw['Low'],errors = 'coerce')
df_raw['Close*'] = pd.to_numeric(df_raw['Close*'],errors = 'coerce')
df_raw['Adj Close**'] = pd.to_numeric(df_raw['Adj Close**'],errors = 'coerce')
df_raw['Volume'] = pd.to_numeric(df_raw['Volume'],errors = 'coerce')
```

```python
print(df_raw)
df_raw.info()
```

```python
# Prophet의 형식이 'ds', 'y'이므로 해당 컬럼으로 데이터 프레임 만들기
df_close_ts = pd.DataFrame({
    'ds' : df_raw['Date'],
    'y' : df_raw['Close*']
})
```

```python
df_close_ts.head()
```

```python
# 연간 계절성 고려
model = Prophet()
model.yearly_seasonality=True
model.daily_seasonality=True
model.fit(df_close_ts);
```

```python
# 30일이 추가된 데이터 프레임 생성
future = model.make_future_dataframe(periods=30)
future.head()
```

```python
forecast = model.predict(future)
forecast.head()
```

```python
# 해당 날짜에 값을 yhat으로 예측
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
model.plot_components(forecast);
```

```python
# 실제값으로 시각화
plt.figure(figsize=(12, 6))
plt.plot(df_close_ts['ds'], df_close_ts['y'], label='real')
plt.plot(forecast['ds'], forecast['yhat'], label='forecast')
plt.grid(True)
plt.legend()
plt.show()
```

```python
model.plot(forecast);
```

```python
# 기아 자동차의 종목코드와 기간을 입력하여 데이터 가져오기
import yfinance as yf
from pandas_datareader import data 

yf.pdr_override()

start_date = "2010-03-01"
end_date = "2018-02-28"
KIA = data.get_data_yahoo("000270.KS", start_date, end_date)
```

```python
KIA.head()
```

```python
KIA.info()
```

```python
KIA['Close'].plot(figsize=(16, 8), grid=True)
```

```python
# 모델의 정확도를 위해 일부 값을 떼어놓고 모델 적합
KIA_trunc = KIA[:"2017-11-30"]
KIA_trunc.tail()
```

```python
df_kia = pd.DataFrame({
    'ds' : KIA_trunc.index, 
    'y' : KIA_trunc['Close']
})
df_kia.reset_index(inplace=True)
del df_kia["Date"]
```

```python
# 인덱스의 값이 들어가 있으므로 이를 제거
df_kia
```

```python
model = Prophet()
model.yearly_seasonality=True
model.weekly_seasonality=True
model.daily_seasonality=True
model.fit(df_kia)
```

```python
future = model.make_future_dataframe(periods=90)
forecast = model.predict(future)
forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]].tail()
```

```python
model.plot(forecast);
```

```python
# 실제값과 비교
plt.figure(figsize=(12, 4))
plt.plot(KIA.index, KIA['Close'], label='real')
plt.plot(forecast['ds'], forecast['yhat'], label = 'forecast')
plt.grid(True)
plt.legend()
plt.show()
```
